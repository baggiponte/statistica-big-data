```python
def solo_pari_tra(input_list, left, right):
    """Definisci una funzione che prende in input una lista e restituisce una nuova lista,
    i cui elementi sono tutti i numeri pari compresi tra due estremi
    
    Example:
    -------
    >>> lista = [1,2,3,4,5,6,7,8,9]
    >>> solo_pari_tra(lista, 2, 7)
        [4,6]
    """
    
    output_list = []
    
    for element in input_list[left:right]:
        if element % 2 == 0:
            output_list.append(element)
    
    return output_list


test_list = [1,2,3,4,5,6,7,8,9]
print(solo_pari_tra(test_list, 2, 5))
```

```python
def elemento_a_meta(lista):
    """Data una lista, restituisce l'elemento a met della lista
    (se la lista  pari, restituire solo uno dei due interi pi vicini).
    
    Example:
    --------
    >>> lista = [1,2,3,4,5]
    >>> elemento_a_meta(lista)
        2
    >>> lista_2 = [1,2,3,4]
    >>> elemento_a_meta(lista_2)
        2
    """
    
    lunghezza = len(lista) // 2
    
    return lista[lunghezza]


lista = [1,2,3,4,5]
elemento_a_meta(lista)
```

```python
def lettera_maiuscola(parola, indice):
    """Ritorna la lettera in posizione `indice` della parola, ma maiuscola
    
    Example:
    --------
    >>> parola = "ciaone"
    >>> lettera_maiuscola(parola, 2)
    A
    """
    
    return parola[indice].upper()


rendi_maiuscola("ciaociao", 3)
```

```python
def rendi_maiuscola(parola, indice):
    
    lettera_maiuscola = parola[indice].upper()
    
    nuova_parola = parola[:indice] + lettera_maiuscola + parola[indice + 1 :]
    
    return nuova_parola


rendi_maiuscola("gattini", 2)
```
